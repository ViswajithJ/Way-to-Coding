```python
import seaborn as sns
```

# **Distribution plots**

```python
tips_df = sns.load_dataset('tips')
```

```python
tips_df.head()
```

```python
# histogram
sns.displot(tips_df['total_bill'],kde=False)
```

```python
# density plot
sns.displot(tips_df['total_bill'],kde=True)
```

```python
sns.displot(tips_df['total_bill'],kde=True,bins=30)
# bins -  the number of bars
```

```python
# default  - scatter plotd
sns.jointplot(x='total_bill',y='tip',data=tips_df)
```

```python
# hex plot
sns.jointplot(x='total_bill',y='tip',data=tips_df,kind='hex')
```

```python
# regression plot inside
sns.jointplot(x='total_bill',y='tip',data=tips_df,kind='reg')
```

```python
# Kernel Distribution Estimation Plot
sns.jointplot(x='total_bill',y='tip',data=tips_df,kind='kde')
```

```python
sns.pairplot(tips_df)
```

```python
sns.pairplot(tips_df,hue='sex')
```

```python
sns.rugplot(tips_df['total_bill'])
```

# **Categorical plots**

```python
sns.countplot(x='sex',data=tips_df)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips_df,hue='smoker')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips_df)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips_df)
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips_df)
```

```python
sns.catplot(x='day',y='total_bill',data=tips_df,kind='bar')
```

# **Matrix plots**

```python
flights_df = sns.load_dataset('flights')
flights_df.head(3)
```

```python
tc = tips_df.corr()
print(tc)
```

```python
sns.heatmap(tc,annot=True)
```

```python
fp = flights_df.pivot_table(index='month',columns='year',values='passengers')
print(fp)
```

```python
sns.heatmap(fp)
```

```python
sns.clustermap(fp)
```

# **Grids**

```python
iris_df = sns.load_dataset('iris')
iris_df.head(3)
```

```python
iris_df['species'].unique()
```

```python
import matplotlib.pyplot as plt
pg = sns.PairGrid(iris_df)
pg.map(plt.scatter)


pg.map_diag(sns.distplot)
pg.map_upper(plt.scatter)
pg.map_lower(sns.kdeplot)
```

```python
fg = sns.FacetGrid(data=tips_df,col='time',row='smoker')
fg.map(sns.distplot,'total_bill')
```
