```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
np.random.seed(101)
data = np.random.randn(5,4)
index = ['A','B','C','D','E']
column = ['W','X','Y','Z']
df = pd.DataFrame(data,index,column)
```

```python
df['W'].hist(bins=20)
```

```python
df['W'].plot(kind='hist')
```

```python
df['W'] = abs(df['W'])
df['W'].plot.area()
```

```python
df.T.plot.line(y='A')
```

```python
df.T.plot.scatter(x='A',y='B',s=df.iloc[0]*100)
```

```python
df.plot.box()
```

```python
df.plot.hexbin(x='W',y='X',gridsize=25)
```
