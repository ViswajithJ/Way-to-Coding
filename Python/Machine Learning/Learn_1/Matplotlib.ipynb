```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# simple graph
x = np.linspace(0,5,11)
y = x ** 2
plt.plot(x,y)
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('Title')
```

```python
# row,column,figure number
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'r')
```

```python
# object oriented method
fig = plt.figure()
# left, bottom, width, and height of the axes
axes = fig.add_axes([0.1,0.1,0.3,0.3])
axes.plot(x,y)
axes.set_xlabel('X label')
axes.set_ylabel('Y label')
axes.set_title('Title')
```

```python
fig = plt.figure()

axes_1 = fig.add_axes([0.1,0.1,0.4,0.4])
axes_2 = fig.add_axes([0.2,0.2,0.2,0.2])

axes_1.plot(x,y)
axes_2.plot(y,x)
```

```python
# subplots
fig,axes = plt.subplots(nrows=1,ncols=2)
#axes[0].plot(x,y)
#axes[1].plot(y,y)
for ax in axes:
  ax.plot(x,y)

# Adjust the padding between and around subplots
plt.tight_layout()
```

```python
# figure size and DPI
fig = plt.figure(figsize=(3,2),dpi=100)
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
# save figure
fig.savefig('graph.png')
```

```python
# legends
fig = plt.figure(figsize=(3,2),dpi=100)
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,label='X')
ax.plot(y,x,label='X')

ax.legend(loc=0)
```

```python
# plot appearance
fig = plt.figure(figsize=(3,2),dpi=100)
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='red',
        linewidth=1,
        alpha=0.5,
        linestyle='-',
        marker='o',
        markersize=10,
        markerfacecolor='white',
        markeredgewidth=3,
        markeredgecolor='blue')

ax.set_xlim([0,5])
ax.set_ylim([0,35])
```
